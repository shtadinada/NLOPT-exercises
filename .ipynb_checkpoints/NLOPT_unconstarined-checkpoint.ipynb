# Optimization Methods for Neural Networks using JAX

In this notebook we will:

- Implement and compare **Gradient Descent (GD)** and **Quasi-Newton (BFGS)** on a small neural network.

- Increase network dimensionality and observe why **Quasi-Newton methods** break down and why **Gradient Descent** is preferred.

- Train a small neural network on **MNIST** using **GD** and **ADAM**, and examine convergence of training loss.

We use **JAX** for algorithmic differentiation (AD) and **Optax** for optimizers (ADAM).


```python
import jax
import jax.numpy as jnp
import optax
import matplotlib.pyplot as plt
from functools import partial
from jax import grad, jacfwd, jacrev, random
from jax.flatten_util import ravel_pytree
from scipy.optimize import minimize
from tensorflow.keras.datasets import mnist
```

## Small Neural Network

We define a simple 1-hidden-layer neural network with tanh activation. We use mean squared error on synthetic data so we can analyze optimization behavior.


```python

key = random.PRNGKey(0)

def init_params(key, n_in=1, n_hidden=10, n_out=1):
    k1, k2 = random.split(key)
    params = {
        "W1": random.normal(k1, (n_in, n_hidden)) * 0.1,
        "b1": jnp.zeros(n_hidden),
        "W2": random.normal(k2, (n_hidden, n_out)) * 0.1,
        "b2": jnp.zeros(n_out),
    }
    return params

def forward(params, x):
    h = jnp.tanh(x @ params["W1"] + params["b1"])
    return h @ params["W2"] + params["b2"]

def loss_fn(params, x, y):
    preds = forward(params, x)
    return jnp.mean((preds - y)**2)

# synthetic data
xs = jnp.linspace(-1, 1, 200).reshape(-1, 1)
ys = xs**3  # nonlinear target
params0 = init_params(key)
```

## Gradient Descent

We implement simple GD using a fixed learning rate (step sizes).

```python
def gradient_descent(params, lr=0.1, steps=100):
    losses = []
    p = params
    for _ in range(steps):
        g = grad(loss_fn)(p, xs, ys)
        p = jax.tree_util.tree_map(lambda w, dw: w - lr * dw, p, g)
        losses.append(loss_fn(p, xs, ys))
    return p, jnp.array(losses)

gd_params, gd_losses = gradient_descent(params0)
```

## Quasi-Newton (BFGS)

We use SciPy's BFGS implementation for convenience.

```python
w0, unravel = ravel_pytree(params0)

# Flattened loss function for SciPy
def loss_flat(w):
    params = unravel(w)
    return float(loss_fn(params, xs, ys))  # convert JAX DeviceArray -> float for SciPy

# Flattened gradient function
def grad_flat(w):
    params = unravel(w)
    g = jax.grad(loss_fn)(params, xs, ys)
    g_flat, _ = ravel_pytree(g)
    return jnp.array(g_flat)

# Store losses manually at each iteration
bfgs_losses = []

# Callback function to record loss during optimization
def callback(w):
    bfgs_losses.append(loss_flat(w))

# Run BFGS using SciPy
res = minimize(
    fun=loss_flat,
    x0=w0,
    jac=grad_flat,
    method="BFGS",
    callback=callback,
    options={"maxiter": 100, "disp": False}
)

# Unravel final optimized parameters
bfgs_params = unravel(res.x)
bfgs_losses = jnp.array(bfgs_losses)
```

## Convergence Comparison

We compare losses of GD, BFGS, and Newton methods on the small network.


```python
plt.plot(gd_losses, label="GD")
plt.plot(bfgs_losses, label="BFGS")
plt.yscale("log")
plt.legend()
plt.title("Loss Convergence (small NN)")
plt.show()
```

# Scaling Network Size

We now increase network width and input dimension.  
Newton and Quasi-Newton methods will become computationally infeasible:

- Newton requires computing a Hessian of size **(params Ã— params)**  
- BFGS stores dense approximations of the Hessian, which is $\mathcal{O}(n^2)$ or worse in memory and time

We only show the failure when attempting larger dimensions. To trains large LLMs such as ChatGPT 4.1, the input dimension is of the order $10^6$.


```python

try:
    # Small/Medium network
    large_params = init_params(key, n_in=50, n_hidden=200, n_out=10)
    w0, _ = ravel_pytree(large_params)
    print("Parameter count:", w0.size)

    # Initialize dense inverse Hessian approximation for BFGS
    H = jnp.eye(w0.size, dtype=jnp.float32)
    print("BFGS inverse Hessian initialized, shape:", H.shape)

    # Try one BFGS step with dummy gradient
    g = jnp.ones_like(w0)  # dummy gradient
    p = -H @ g
    w_new = w0 + 0.1 * p
    print("One BFGS step computed successfully")

except Exception as e:
    print("BFGS failed due to large dimension:", e)

```

# MNIST Training with GD and ADAM

We now train a simple MLP on MNIST using:

- Gradient Descent  
- [ADAM](https://arxiv.org/abs/1412.6980) (momentum based method)  

We compare the training loss curves.


```python
# Load MNIST

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(-1, 28*28) / 255.0
x_test  = x_test.reshape(-1, 28*28) / 255.0

y_train = jax.nn.one_hot(y_train, 10)
y_test  = jax.nn.one_hot(y_test, 10)

x_train = jnp.array(x_train)
y_train = jnp.array(y_train)
x_test  = jnp.array(x_test)
y_test  = jnp.array(y_test)
```

```python
# Model
def init_mnist_params(key):
    return {
        "W1": random.normal(key, (784, 128)) * 0.01,
        "b1": jnp.zeros(128),
        "W2": random.normal(key, (128, 10)) * 0.01,
        "b2": jnp.zeros(10),
    }

def mnist_forward(params, x):
    h = jax.nn.relu(x @ params["W1"] + params["b1"])
    return h @ params["W2"] + params["b2"]

def mnist_loss(params, x, y):
    logits = mnist_forward(params, x)
    return optax.softmax_cross_entropy(logits, y).mean()


# Training methods
def train_gd(params, lr=0.1, steps=200):
    losses = []
    for _ in range(steps):
        g = grad(mnist_loss)(params, x_train, y_train)
        params = jax.tree_util.tree_map(lambda w, dw: w - lr * dw, params, g)
        losses.append(mnist_loss(params, x_train, y_train))
    return params, jnp.array(losses)

def train_adam(params, lr=0.001, batch_size=128, steps=200):
    opt = optax.adam(lr)
    opt_state = opt.init(params)

    losses = []
    key = random.PRNGKey(1)

    for _ in range(steps):
        key, sub = random.split(key)
        idx = random.choice(sub, len(x_train), shape=(batch_size,), replace=False)
        xb, yb = x_train[idx], y_train[idx]

        loss, grads = jax.value_and_grad(mnist_loss)(params, xb, yb)
        updates, opt_state = opt.update(grads, opt_state)
        params = optax.apply_updates(params, updates)

        losses.append(loss)

    return params, jnp.array(losses)

# Accuracy function
def mnist_accuracy(params, x, y):
    logits = mnist_forward(params, x)
    preds = jnp.argmax(logits, axis=1)
    labels = jnp.argmax(y, axis=1)
    correct = jnp.sum(preds == labels)
    accuracy = correct / x.shape[0]
    return int(correct), float(accuracy)
```

```python
# Run training
mnist_params = init_mnist_params(random.PRNGKey(1))

gd_params, gd_train_loss = train_gd(mnist_params)
adam_params, adam_loss = train_adam(mnist_params)

# Compute accuracy
gd_correct, gd_acc = mnist_accuracy(gd_params, x_test, y_test)
adam_correct, adam_acc = mnist_accuracy(adam_params, x_test, y_test)

print(f"GD correct:   {gd_correct} / {len(x_test)}   ({gd_acc*100:.2f}%)")
print(f"Adam correct: {adam_correct} / {len(x_test)}   ({adam_acc*100:.2f}%)")

# Plot losses
plt.plot(gd_train_loss, label="GD")
plt.plot(adam_loss, label="Adam")
plt.yscale("log")
plt.legend()
plt.title("MNIST Training Loss")
plt.show()
```
